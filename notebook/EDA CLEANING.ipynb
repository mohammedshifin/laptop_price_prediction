```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Memory'] = encoder.fit_transform(df['Memory'])
df['Product'] = encoder.fit_transform(df['Product'])
```

```python
df['ppi'] = (((df['X_res']**2) + (df['Y_res']**2))**0.5/df['Inches']).astype('float')
```

```python
df['GPU_Company'].value_counts()
```

```python
df.drop(columns = 'GPU_Type' ,inplace=True)
```

```python
df = df[df['GPU_Company'] !='ARM']
```

```python
df['OpSys'].value_counts()
```

```python
def cat_os(inp):
    if inp == 'Windows 10' or inp == 'Windows 7' or inp == 'Windows 10 S':
        return 'Windows'
    elif inp == 'macOS' or inp == 'Mac OS X':
        return 'Mac'
    else:
        return 'Others/No OS/Linux'
```

```python
df['os'] = df['OpSys'].apply(cat_os)
```

```python
df.drop(columns=['OpSys','CPU_Type','ScreenResolution','X_res','Y_res',"Inches"],inplace=True)
```
